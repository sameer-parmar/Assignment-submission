```python
import matplotlib.pyplot as plt

def find_non_intersecting_paths(flights):
    paths = []

  
    
    for flight in flights:
        path = []
        for point in flight:
            path.append(tuple(map(int, point.split(','))))
        paths.append(path)
    

    def segments_intersect(seg1, seg2):
        def orientation(p, q, r):
            val = (q[1] - p[1]) * (r[0] - q[0]) - (q[0] - p[0]) * (r[1] - q[1])
            if val == 0:
                return 0  # Collinear
            return 1 if val > 0 else 2 

        p1, q1 = seg1
        p2, q2 = seg2
        o1 = orientation(p1, q1, p2)
        o2 = orientation(p1, q1, q2)
        o3 = orientation(p2, q2, p1)
        o4 = orientation(p2, q2, q1)

        if o1 != o2 and o3 != o4:
            return True
        return False

   
    adjusted_paths = []
    for path in paths:
        new_path = []
        for point in path:
            intersecting = False
            if new_path:
                for other_path in adjusted_paths:
                    for i in range(len(other_path) - 1):
                        if segments_intersect((new_path[-1], point), (other_path[i], other_path[i + 1])):
                            intersecting = True
                            break
                    if intersecting:
                        break
                if intersecting:
                   
                    point = (point[0] + 1, point[1] + 1)
            new_path.append(point)
        adjusted_paths.append(new_path)

    return adjusted_paths

def plot_flight_paths(paths):
    colors = ['r', 'g', 'b', 'c', 'm', 'y', 'k']
    for i, path in enumerate(paths):
        x_coords, y_coords = zip(*path)
        plt.plot(x_coords, y_coords, marker='o', color=colors[i % len(colors)], label=f'Flight {i+1}')
    
    plt.xlabel('X Coordinate')
    plt.ylabel('Y Coordinate')
    plt.title('Flight Paths')
    plt.legend()
    plt.grid(True)
    plt.show()

flights = [
    ["1,1", "2,2", "3,3"],
    ["1,1", "2,4", "3,2"],
    ["1,1", "4,2", "3,4"]
]

non_intersecting_paths = find_non_intersecting_paths(flights)
for path in non_intersecting_paths:
    print("Path:", path)

plot_flight_paths(non_intersecting_paths)
```

```python
def is_valid_move(current, next_position):
    return next_position[1] < current[1] and (next_position[0] == current[0] or next_position[1] == current[1])

def find_paths(castle, soldiers, path, visited):
    if len(path) == len(soldiers) + 1:
        if is_valid_move(path[-1], castle):  # Check if it can return to start
            return [path + [castle]]
        return []
    
    paths = []
    for i, soldier in enumerate(soldiers):
        if not visited[i]:
            visited[i] = True
            new_path = path + [soldier]
            new_paths = find_paths(castle, soldiers, new_path, visited)
            paths.extend(new_paths)
            visited[i] = False
    return paths

def print_path(path):

    print("Start:", path[0])
    for i in range(1, len(path)):
        if i == len(path) - 1:
            print(f"Arrive {path[i]}")
        else:
            direction = "Left"
            print(f"Kill {path[i]}. Move {direction}")

def main():
    try:
        soldier_count = int(input("Enter the number of soldiers: "))
        if soldier_count <= 0:
            raise ValueError("Number of soldiers must be a positive integer.")
    except ValueError as e:
        print(f"Invalid input for soldier count: {e}")
        return
    
    soldiers = []
    for i in range(soldier_count):
        while True:
            try:
                x, y = map(int, input(f"Enter coordinates for soldier {i + 1} (x,y): ").split(','))
                soldiers.append((x, y))
                break
            except ValueError:
                print("Invalid input. Please enter coordinates in 'x,y' format.")
    
    while True:
        try:
            castle_x, castle_y = map(int, input("Enter the coordinates for your 'special' castle (x,y): ").split(','))
            castle = (castle_x, castle_y)
            break
        except ValueError:
            print("Invalid input. Please enter coordinates in 'x,y' format.")
    
    visited = [False] * len(soldiers)
    paths = find_paths(castle, soldiers, [castle], visited)
    
    if paths:
        print(f"Thanks. There are {len(paths)} unique paths for your 'special castle':")
        for i, path in enumerate(paths, 1):
            print(f"Path {i}")
            print("=======")
            print_path(path)
            print()
    else:
        print("No path found that satisfies the conditions.")

if __name__ == "__main__":
    main()
```

```python
def distribute_apples():
 
    apple_weights = []
    while True:
        try:
            weight = int(input("Enter apple weight in grams (-1 to stop): "))
            if weight == -1:
                break
            elif weight <= 0:
                print("Please enter a positive weight.")
            else:
                apple_weights.append(weight)
        except ValueError:
            print("Invalid input. Please enter a number.")
    
    if not apple_weights:
        print("No apple weights were entered.")
        return
    
 
    apple_weights.sort(reverse=True)
    
   
    total_weight = sum(apple_weights)
    

    ideal_weight_ram = total_weight * 0.50
    ideal_weight_sham = total_weight * 0.30
    ideal_weight_rahim = total_weight * 0.20
    
    print(f"Total weight of apples: {total_weight}g")
    print(f"Ideal weight for Ram: {ideal_weight_ram}g")
    print(f"Ideal weight for Sham: {ideal_weight_sham}g")
    print(f"Ideal weight for Rahim: {ideal_weight_rahim}g")
    
    
    ram_apples = []
    sham_apples = []
    rahim_apples = []
    
    ram_weight = 0
    sham_weight = 0
    rahim_weight = 0
    
    for weight in apple_weights:
        if ram_weight < ideal_weight_ram:
            ram_apples.append(weight)
            ram_weight += weight
        elif sham_weight < ideal_weight_sham:
            sham_apples.append(weight)
            sham_weight += weight
        elif rahim_weight < ideal_weight_rahim:
            rahim_apples.append(weight)
            rahim_weight += weight
        else:
            
            if ram_weight < ideal_weight_ram:
                ram_apples.append(weight)
                ram_weight += weight
            elif sham_weight < ideal_weight_sham:
                sham_apples.append(weight)
                sham_weight += weight
            else:
                rahim_apples.append(weight)
                rahim_weight += weight
    
   
    print("\nDistribution Result:")
    print(f"Ram: {', '.join(map(str, ram_apples))} (total {ram_weight}g, {ram_weight/total_weight:.0%})")
    print(f"Sham: {', '.join(map(str, sham_apples))} (total {sham_weight}g, {sham_weight/total_weight:.0%})")
    print(f"Rahim: {', '.join(map(str, rahim_apples))} (total {rahim_weight}g, {rahim_weight/total_weight:.0%})")


distribute_apples()
```
